# series

```python
import pandas as pd
```

```python
x=[1,7,5,3,9,6]

var = pd.Series(x)
print(var)
print(type(var))
print(var[2])
```

```python
x=[1,7,5]

var = pd.Series(x,dtype='float',index=['a','h','l'])
print(var)
print(type(var))
print(var[2])
```

```python
s=pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
s1=pd.Series(12,index=[1,2,3,4,5,6,7])
s2=pd.Series(12,index=[1,2,3,4,5])
var1=s1+s2
print(var1)
print(type(var1))
```

# DATAFRAME

```python
l=[1,2,3,4,5]
var1= pd.DataFrame(l)
print(var1)
print(type(var1))
```

```python
d={'no':[1,3,4,5],'subject':['english','math','coms','phyisc']}
var1=pd.DataFrame(d)

print(d)
print(var1)
print(type(d))
```

```python
d={'no':[1,3,4,5],'subject':['english','math','coms','phyisc'],'a':['mm','mn','kl','ok']}
var1=pd.DataFrame(d)

print(d)
print(var1)
print(type(d))
print(var1["no"][3])
```

# WRITE CSV

```python
dis= pd.DataFrame({"a":[1,3,4,5,6],"s":[1,4,6,9,7],"d":[1,3,4,5,9]})
dis

d= pd.DataFrame(dis)
print(d)

d.to_csv("test_new_csv")
```
